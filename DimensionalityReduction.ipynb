```python
import numpy as np 
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pylab as plt
```

```python
iris = datasets.load_iris()
X = iris.data  # we only take the first two features.
y = iris.target

Xh = PCA(n_components=2).fit_transform(X)

plt.figure(dpi=500)
plt.plot(Xh[y==0, 0], Xh[y==0, 1], 'o', label='setosa')
plt.plot(Xh[y==1, 0], Xh[y==1, 1], 'o', label='versicolor')
plt.plot(Xh[y==2, 0], Xh[y==2, 1], 'o', label='virginica')
plt.legend()
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.savefig('pdf/dimred_pca_iris.pdf')
```

```python
D = datasets.load_wine()
X = D.data  # we only take the first two features.
y = D.target


Xh = PCA(n_components=2).fit_transform(StandardScaler().fit_transform(X))

plt.figure(dpi=500)
plt.plot(Xh[y==0, 0], Xh[y==0, 1], 'o', label='Wine0')
plt.plot(Xh[y==1, 0], Xh[y==1, 1], 'o', label='Wine1')
plt.plot(Xh[y==2, 0], Xh[y==2, 1], 'o', label='Wine2')
plt.legend()
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.savefig('pdf/dimred_pca_wine.pdf')

Xh = PCA(n_components=2).fit_transform(X)

plt.figure(dpi=500)
plt.plot(Xh[y==0, 0], Xh[y==0, 1], 'o', label='Wine0')
plt.plot(Xh[y==1, 0], Xh[y==1, 1], 'o', label='Wine1')
plt.plot(Xh[y==2, 0], Xh[y==2, 1], 'o', label='Wine2')
plt.legend()
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.savefig('pdf/dimred_pca_wine_nostd.pdf')
```

```python
D = datasets.load_wine()
X = D.data  # we only take the first two features.
y = D.target


pc = PCA().fit(StandardScaler().fit_transform(X))
plt.figure(dpi=500)
plt.plot(np.cumsum(pc.explained_variance_))
plt.legend()
plt.xlabel('PC')
plt.ylabel('Variance Explained')
plt.savefig('pdf/dimred_pca_variance_wine.pdf')
```

```python
from sklearn.datasets import load_digits
from sklearn.manifold import MDS
X, y = load_digits(return_X_y=True)
embedding = MDS(n_components=2)
X_transformed = embedding.fit_transform(X[:1000])
```

```python
yy = y[:1000]
plt.figure(dpi=500)
for i in range(10): 
    plt.plot(X_transformed[yy==i, 0], X_transformed[yy==i, 1], 'o', label=str(i))
plt.legend()
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.savefig('pdf/dimred_mds_mnist.pdf')
```
